```python
#Finding and replacing outliers
```

```python
sns.boxplot(cc.CreditScore)
```

```python
q1=cc.CreditScore.quantile(0.25)
q3=cc.CreditScore.quantile(0.75)
print(q1)
print(q3)
```

```python
iqr=q3-q1
iqr
```

```python
ll=q1-1.4*iqr
ll
```

```python
cc.CreditScore.median()
```

```python
cc['CreditScore']=np.where(cc['CreditScore']<ll,600.0,cc['CreditScore'])
```

```python
sns.boxplot(cc.CreditScore)
```

```python
sns.boxplot(cc.NumOfProducts)
```

```python
q1=cc.NumOfProducts.quantile(0.23)
q3=cc.NumOfProducts.quantile(0.65)
print(q1)
print(q3)
```

```python
IQR=q3-q1
IQR
```

```python
UL=q3+1.5*IQR
UL
```

```python
cc.NumOfProducts.median()
```

```python
cc['NumOfProducts']=np.where(cc['NumOfProducts']>UL,1.0,cc['NumOfProducts'])
```

```python
sns.boxplot(cc.NumOfProducts)
```
