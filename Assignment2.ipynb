```python
#Finding and replacing outliers
```

```python
sns.boxplot(df.CreditScore)
```

```python
q1=df.CreditScore.quantile(0.25)
q3=df.CreditScore.quantile(0.75)
print(q1)
print(q3)
```

```python
iqr=q3-q1
iqr
```

```python
ll=q1-1.4*iqr
ll
```

```python
df.CreditScore.median()
```

```python
df['CreditScore']=np.where(df['CreditScore']<ll,600.0,df['CreditScore'])
```

```python
sns.boxplot(df.CreditScore)
```

```python
sns.boxplot(df.NumOfProducts)
```

```python
q1=df.NumOfProducts.quantile(0.23)
q3=df.NumOfProducts.quantile(0.65)
print(q1)
print(q3)
```

```python
IQR=q3-q1
IQR
```

```python
UL=q3+1.5*IQR
UL
```

```python
df.NumOfProducts.median()
```

```python
df['NumOfProducts']=np.where(df['NumOfProducts']>UL,1.0,df['NumOfProducts'])
```

```python
sns.boxplot(df.NumOfProducts)
```
