```python
#Finding and replacing outliers
sns.boxplot(fd.CreditScore)
```

```python
q1=fd.CreditScore.quantile(0.25)
q3=fd.CreditScore.quantile(0.75)
print(q1)
print(q3)
```

```python
iqr=q3-q1
iqr
```

```python
ll=q1-1.5*iqr
ll
```

```python
fd.CreditScore.median()
```

```python
fd['CreditScore']=np.where(fd['CreditScore']<ll,652.0,fd['CreditScore'])
```

```python
sns.boxplot(fd.CreditScore)
```

```python
sns.boxplot(fd.NumOfProducts)
```

```python
Q1=fd.NumOfProducts.quantile(0.25)
Q3=fd.NumOfProducts.quantile(0.75)
print(Q1)
print(Q3)
```

```python
IQR=Q3-Q1
IQR
```

```python
UL=Q3+1.5*IQR
UL
```

```python
fd.NumOfProducts.median()
```

```python
fd['NumOfProducts']=np.where(fd['NumOfProducts']>UL,1.0,fd['NumOfProducts'])
```

```python
sns.boxplot(fd.NumOfProducts)
```
